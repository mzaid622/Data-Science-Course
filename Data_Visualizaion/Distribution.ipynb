```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Distribution Plots in Seaborn  

We use **distribution plots** to visualize the distribution of quantitative data.  
Some useful plots in Seaborn are:  

1. **distplot / histplot** → Shows histogram with KDE (Kernel Density Estimate).  
2. **jointplot** → Displays bivariate distribution with scatter + histograms.  
3. **pairplot** → Shows pairwise relationships between multiple variables.  
4. **rugplot** → Places small vertical ticks along the axis to show data points.  


```python
# Hisplot
df=sns.load_dataset("tips")
```

```python
df
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins=10,kde=True)
```

```python
# Joint Plot
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
# Pair Plot
sns.pairplot(df,hue="sex",palette="rainbow")
```

```python
sns.pairplot(df,hue="size",palette="rainbow")
```
