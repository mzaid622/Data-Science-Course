Pandas DataFrame

```python
import numpy as np
import pandas as pd
```

Creating a DataFrame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
# l1 = ["a","b","c","d"]
df=pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
header = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns=header)
```

```python
df2
```

Selection and Indexing of Columns

```python
df2[["Name","Age"]]
```

Creating a new Column

```python
df2["Designation"] = ["Doc","Egr","Law","Doc"]
```

```python
df2
```

Remove a Column

```python
# This will not permentally delete data in memory we use one attribute
df2.drop("Designation",axis=1)
```

```python
df2.drop("Designation",axis=1,inplace=True)
```

```python
df2
```

Selecting Rows

```python
df2.loc[0:2]
```

```python
df2.iloc[0:2]
```

Selecting Subsets of Row and Columns

```python
df2.iloc[0:2,2:4]
```

```python
# 2nd Method
df2.loc[[0,1]][["City","Salary"]]
```

```python
# I want those who age is above 30 years
df2[df2["Age"] > 30]
```
