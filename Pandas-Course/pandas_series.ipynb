Pandas Series

```python
import numpy as np
import pandas as pd
```

Creating Pandas Series

```python
labels = ["a","b","c"]
my_list = [1,2,3]
arr = np.array([1,2,3])
dictionary = {1:10,2:20,3:30}
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(dictionary)
```
