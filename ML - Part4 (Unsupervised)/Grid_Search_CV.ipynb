```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
df['species'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X= df.drop('species',axis= 1)
y= df['species']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(X_train,Y_train)
```

```python
knn_model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC

model_svc = SVC(gamma='auto')
```

```python
model_svc.fit(X_train,Y_train)
```

```python
model_svc.score(X_test,Y_test)
```

```python
#now lets use the grid search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svc),{
  'C':[1,10,20,30],
  'kernel' :['rbf','linear']
},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
res=pd.DataFrame(classifier.cv_results_)
```

```python
res.head(8)
```

```python
res[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
```

```python
knn_classifer = GridSearchCV(model_knn,
{
  'n_neighbors': [3,5,7,9,11,13],
  'weights':['uniform','distance'],
  'metric':['euclidean','manhattan']
  },cv=5,return_train_score=False
)
```

```python
knn_classifer.fit(X,y)
```

```python
knn_classifer.cv_results_
```

```python
knn_res = pd.DataFrame(knn_classifer.cv_results_)
```

```python
knn_res.head()
```

```python
knn_res[['param_n_neighbors','param_weights','params','mean_test_score']].sort_values(by='mean_test_score',ascending=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svc),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},n_iter = 4,cv = 5,return_train_score = False)
```

```python
classifier_r.fit(X,y)
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```
