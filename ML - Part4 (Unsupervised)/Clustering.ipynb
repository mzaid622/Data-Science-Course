```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=4, random_state=42)
```

```python
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
```

```python
df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
inertia = []
K_range = range(1, 11)
```

```python
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(K_range, inertia, marker='o')
```

```python

kmeans_final = KMeans(n_clusters=3, random_state=42)
```

```python
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
df['cluster'] = cluster_labels
```

```python
sns.scatterplot(x=df['Feature_1'],
                y=df['Feature_2'],
                hue=df['cluster'],
                palette='viridis')
```

```python
from sklearn.datasets import make_moons
```

```python
X,y_true = make_moons(n_samples=500,noise=0.05, random_state=42)
```

```python
from sklearn.cluster import KMeans,DBSCAN
```

```python
df1 = pd.DataFrame(X,columns=['Feature_1','Feature_2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df1)
```

```python
kmeans_final = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
df1['cluster'] = cluster_labels
```

```python
sns.scatterplot(x=df1['Feature_1'],
                y=df1['Feature_2'],
                hue=df1['cluster'],
                palette='viridis')
```

```python
dbscan = DBSCAN(eps=0.3,min_samples=5)
dbscan_lable =dbscan.fit_predict(X_scaled)
```

```python
df1['dbscan_cluster'] = dbscan_lable
```

```python
sns.scatterplot(x=df1['Feature_1'],
                y=df1['Feature_2'],
                hue=df1['dbscan_cluster'],
                palette='viridis')
```
