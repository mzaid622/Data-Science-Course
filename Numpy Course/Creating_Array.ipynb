Random Generation Function

```python
# This function use concept of Normaizaltion always give answer between 0 to 1
np.random.rand(3)
```

```python
# This function use concept of Normaizaltion always give answer between -3 to 3
np.random.randn(5)
```

```python
# This function will print only 1 number between 0 to 5
np.random.randint(5)
```

```python
# if we want more random number
np.random.randint(10,20,10)
```

```python
# If you want unique numbers each time
nums = np.random.choice(np.arange(10, 20), size=10, replace=False)
print(nums)
```
