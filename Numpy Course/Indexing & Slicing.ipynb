```python
import numpy as np
```

Numpy Array Indexing and Slicing of Vector

```python
arr = np.arange(11,21)
arr
```

```python
arr[4]
```

```python
arr[0:5]
```

```python
arr[1::3]
```

```python
arr[-1]
```

Numpy Array Indexing and Slicing of Matrix

```python
arr=np.arange(1,26).reshape(5,5)
arr
```

```python
arr[1][2]
```

```python
arr[0:2,1:3]
# arr[0:2][1:3]
```

```python
arr[2:,3:]
```

```python
arr[:,2]
```

Boolean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr=arr[bool_index]
```
